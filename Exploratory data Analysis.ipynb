```python
numerical_columns_in_original_data = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population', 'Area (kmÂ²)', 'Density (per kmÂ²)', 'Growth Rate', 'World Population Percentage']

for col in numerical_columns_in_original_data:
  if col in data.columns:
    data[col]= pd.to_numeric(data[col], errors='coerce')
```

```python
df_numeric_features = data.select_dtypes(include=np.number).copy()
df_numeric_features.head()
```

```python
categorical_cols_to_encode = ['CCA3', 'Country', 'Capital', 'Continent']
df_categorical_features = data[categorical_cols_to_encode].copy()
df_categorical_features.head()

df_encoded_categorial_features = pd.get_dummies(df_categorical_features, dtype = float)
df_encoded_categorial_features.head(4)
```

```python
final_df_for_correlation = pd.concat([df_numeric_features, df_encoded_categorial_features], axis = 1)
```

```python
print("\n--- Final Destination for Correlation (Head): ---")
print(final_df_for_correlation.head())
```

```python
print("\n--- Final Destination for Info (Check dtypes): ---")
print(final_df_for_correlation.info())
```

```python
correlation_matrix = final_df_for_correlation.corr()



print(correlation_matrix.stack().describe()) # Flatten and get stats

print("\n--- Min/Max/Mean of Correlation Matrix (excluding NaNs) ---")
print(f"Min Correlation: {correlation_matrix.min().min()}")
print(f"Max Correlation: {correlation_matrix.max().max()}")
print(f"Mean Correlation: {correlation_matrix.mean().mean()}")

# Check for all NaNs (should be False if you confirmed it earlier)
print(f"Is correlation_matrix full of NaNs? {correlation_matrix.isnull().all().all()}")
```

```python
if not correlation_matrix.empty and not correlation_matrix.isnull().all().all():
    num_cols = correlation_matrix.shape[1]
    fig_width = max(20, num_cols * 0.5)
    fig_height = max(15, num_cols * 0.4)
    plt.figure(figsize=(fig_width, fig_height))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax = 1)
    plt.title("Correlation Heatmap")
    plt.xticks(rotation=90, fontsize=10)
    plt.yticks(rotation=0, fontsize=10)
    plt.tight_layout()
    plt.show()
else:
    print("Correlation matrix is empty or contains NaN values.")
```
